## Sujet
xmind est un logiciel de mindmapping.     
Il permet de réaliser des cartes heuristiques qui peuvent se voir comme des listes et sur lesquels on peut rajouter des informations complémentaires 
comme des liens, des markers(parmi ceux proposés(etoiles, drapeaux,...)), ou encore des labels (des mots)    
      
Le problème que l'on cherche à résoudre c'est que le contenu d'un carte xmind peut rapidement devenir important et on peine à retrouver les markers que l'on a laissé sur sa carte.      

Un fichier est basé sur un fichier xml avec des noms propres aux fichiers xmind.

Il s'agit ici de lire une carte xmind et non d'écrire un fichier xmind.

## Definitions
- node : On appelle node la serie de nombre entier séparer par '-' qui désigne la numérotation des sous sujets à partir de la racine du fichier.  ". Exemple '0-1-0'     
- topic :  le contenu textuel d'un node    
- marker :  désigné par des chaines de caractères comme , il peut y en avoir plusieurs sur le même node
- labels :  liste de mots 
- link : contient un url

## fonctionnalités 
- lire le fichier à partir d'un node
- on cherche notamment à pouvoir retrouver tous les sujets avec un marker 
- on cherche notamment à pouvoir trouver et compter tous les markers à partir d'un certain endroit de la carte.
